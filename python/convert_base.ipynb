```python
from glob import glob

import matplotlib.pyplot as plt
import numpy as np
from scipy.io import loadmat
import pandas as pd
import os
```

```python
train_csv = pd.read_csv('../training2017/REFERENCE.csv', header=None)
train_csv.columns = ['file', 'label']

files = []
labels = []
for index, row in train_csv.iterrows():
    f = row['file']
    l = row['label']
    if l=="N" or l=="A":
        files.append(f)
        labels.append(l)
        
processed_train = pd.DataFrame(columns=['file', 'label'])
processed_train['file'] = files
processed_train['label'] = labels
processed_train.to_csv('train.csv')
processed_train.shape
```

```python
test_csv = pd.read_csv('../sample2017/validation/REFERENCE.csv', header=None)
test_csv.columns = ['file', 'label']

files = []
labels = []
for index, row in test_csv.iterrows():
    f = row['file']
    l = row['label']
    if l=="N" or l=="A":
        files.append(f)
        labels.append(l)
        
processed_test = pd.DataFrame(columns=['file', 'label'])
processed_test['file'] = files
processed_test['label'] = labels
processed_test.to_csv('test.csv')
processed_test.shape
```
